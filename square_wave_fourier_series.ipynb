Import the required packages.

```python
import pandas as pd
import numpy as np
import math
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
L = 2
m = 1000
x = np.linspace(0, 2*L, m)
#Generate Square Wave
Y = []
for n in x:
    if n <= L:
        y = 1
    else:
        y = -1
    Y.append(y)
#Add fourier components
for n in range(3,35,5):
    print(n, 'Fourier components.')
    #n = 30 # choose the number of Fourier Components to add
    sum = np.zeros(m)
    for i in range(1, n, 2):
        sin = np.sin(i*math.pi*x/L)/i
        sum = sin + sum  
    sum = 4*sum/math.pi
    sns.set(font_scale= 2)
    sns.set_style('whitegrid')
    plt.figure(figsize= (15,7))
    sns.lineplot(x = x, y = sum, linewidth = 5)
    sns.lineplot(x = x, y = Y, color = 'red')
    plt.xlabel('x')
    plt.ylabel('y');
```
